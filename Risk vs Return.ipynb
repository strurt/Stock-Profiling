```python

import pandas as pd
import yfinance as yt
import matplotlib.pyplot as plt
import seaborn
import numpy as np
```

```python
stocks = ["AAPL", "BA", "GE", "GOOG"]
```

```python
yt.download(stocks,start = "2018-01-01",end = "2023-05-05")
```

```python
x = yt.download(stocks,start = "2018-01-01",end = "2023-05-05")
```

```python
df = x.loc[:,"Close"].copy()
```

```python
df.describe()
```

```python
df.pct_change()
```

```python
remove_null = df.pct_change().dropna()
```

```python
remove_null.describe()
```

```python
k = remove_null
```

```python
get = k.describe().T.loc[:, ["mean","std"]]
```

```python
get
```

```python
get["mean"]= get["mean"]* 252
```

```python
get["std"]= get["std"]*np.sqrt(252)
```

```python
get
```

```python
get.plot.scatter(x = "std", y="mean",figsize=(9,6),s=5,fontsize=8)
for i in get.index:
    plt.annotate(i,xy = (get.loc[i,"std"]+0.001,get.loc[i,"mean"]+0.001),size=8)
```

```python
plt.show()
```

```python
rnk = get.copy()
```

```python
rnk['mean_rank'] = get['mean'].rank(ascending=False)
rnk['std_rank'] = get['std'].rank(ascending=False)
```
