```python
import yfinance as yt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn 
```

```python
stocks = ["AAPL", "BA", "GE",  "GOOG"]
```

```python
yt.download(stocks,start = "2015-01-01",end = "2023-05-05")
```

```python
f_stocks = yt.download(stocks,start = "2015-01-01",end = "2023-05-05")
```

```python
df= f_stocks.loc[:,"Close"].copy()
```

```python
df.plot(figsize= (19,9),fontsize = 18 )
plt.xlabel('Year',fontsize = 18)
plt.ylabel('Price',fontsize=18)
```

```python
plt.style.use("seaborn")
```
